```python
import numpy as np
import pandas as pd
```

```python
data= {'NAME':['abc','XYX','pqr','YZX','MNO','CAB'],
'GENDER' :['M', 'F', 'F','M', 'F','F'],
'AGE':[26,np.NaN,21,np.NaN,21,19],
'MARKS' : [75,56,np.NaN,56,np.NaN,67]
}
```

```python
df=pd.DataFrame(data)
df.head()
```

```python
df.isnull()
```

```python
data2= {'NAME':['abc','XYX','pqr','YZX','MNO','CAB'],
'GENDER':['M','F','F','M','F','F'],
'AGE':[26,np.NaN,21,np.NaN,21,19],
'MARKS':[75,56,np.NaN,56,np.NaN,67],
'PHD':['Y','N','N','Y','13','N']
}
df2=pd.DataFrame(data2)
df2
```
