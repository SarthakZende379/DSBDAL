```python
def func(Age):
    if (Age in range(20,25)):
        return 0
    elif (Age in range(25,30)):
        return 1
    elif (Age in range(30,35)):
        return 2
    else:
        return 3
```

```python
df['Age']=df['Age'].apply(func)
```
